# Plots

The goal in this notebook is to visualize the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
from matplotlib import animation
```

```python
df = pd.read_csv("cleaned_data.csv")

df
```

```python
X_Time = df["Time"].to_numpy().reshape(-1,1)
X_Cases = df["Cases"].to_numpy().reshape(-1,1)
X_Close = df["Close"].to_numpy().reshape(-1,1)

y = df["next_Close"].to_numpy()
```

```python
# Plot the features against label and time against cases
fig, axes = plt.subplots(2,2, figsize= (12,10))

axes[0][0].scatter(X_Time,y)
axes[0][0].set_title("Time vs next_Close")
axes[0][0].set_xlabel("Time")
axes[0][0].set_ylabel("next_Close")

axes[0][1].scatter(X_Cases,y)
axes[0][1].set_title("Cases vs next_Close")
axes[0][1].set_xlabel("Cases")
axes[0][1].set_ylabel("next_Close")

axes[1][0].scatter(X_Close,y)
axes[1][0].set_title("Close vs next_Close")
axes[1][0].set_xlabel("Close")
axes[1][0].set_ylabel("next_Close")

axes[1][1].scatter(X_Time,X_Cases)
axes[1][1].set_title("Time vs Cases")
axes[1][1].set_xlabel("Time")
axes[1][1].set_ylabel("Cases")

```

```python
## Plot features against label

# Make plot interactive
%matplotlib widget 

fig1 = plt.figure(figsize=(12, 6))
ax = fig1.add_subplot(projection="3d")

# Choose not to include "Close" value plot since it is much correlated with the label
# and we have to leave one feature out since we are constrained by the dimensions
ax.scatter(X_Time, X_Cases, y, s=10)
ax.set_xlabel('Time')
ax.set_ylabel('Cases')
ax.set_zlabel('next_Close')

```
